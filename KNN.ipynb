# Feature Selections
It is the process of reducing the number of input variables when developing a predictive model. Used to reduce the number of input variables to both reduce the computational cost of modelling and, in some cases, to improve the performance of the model.

````python
features = df.iloc[:, :-1]  # Selecting all columns except the last one as features
labels = df.iloc[:, -1]    # Selecting the last column as labels
target = df['label']
````

# Data Splitting
The train-test split is a procedure for evaluating the performance of an algorithm. The procedure involves taking a dataset and dividing it into two subsets. The first subset is utilized to fit/train the model. The second subset is used for prediction. The main motive is to estimate the performance of the model on new data.

````python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=2)

# Printing the number of samples in each set
print("Number of samples in the training set:", len(X_train))
print("Number of samples in the testing set:", len(X_test))
````

## K-Nearest Neighbors










````python
from sklearn.metrics import confusion_matrix
def plot_conf_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()
# K-Nearest Neighbors

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print("K-Nearest Neighbors Report:\n", classification_report(y_test, y_pred_knn))
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
plot_conf_matrix(y_test, y_pred_knn, 'K-Nearest Neighbors')

````

# Overfitting in K-Nearest Neighbors
Overfitting occurs when a model performs well on the training data but poorly on unseen test data. This typically happens when the model captures noise and details in the training data to an extent that it negatively impacts the performance on new data.

##  Symptoms of Overfitting
In the context of the K-Nearest Neighbors (KNN) classifier:

High accuracy on training data but low accuracy on test data.
The model might be too sensitive to the specific instances in the training set.
#Possible Reasons for Overfitting in KNN
The resasons  is that over data ha most of the value of the rise they tell most of the prediction that their is rise
# Feature Scaling:
Inadequate feature scaling can cause certain features to dominate others, leading to suboptimal performance.
High Dimensionality: High-dimensional data can make the distance metrics less informative, leading to overfitting.
# Diagnosing Overfitting
To diagnose overfitting, we can compare the model's performance on the training data and the test data.

## Example Code for Evaluating Overfitting

````python
# K-Nearest Neighbors with k=1 (likely to overfit)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Predictions on training data
y_train_pred_knn = knn.predict(X_train)
# Predictions on test data
y_test_pred_knn = knn.predict(X_test)

# Evaluating on training data
print("KNN Classifier (k=1) Report on Training Data:\n", classification_report(y_train, y_train_pred_knn))
print("Training Accuracy:", accuracy_score(y_train, y_train_pred_knn))

# Evaluating on test data
print("KNN Classifier (k=1) Report on Test Data:\n", classification_report(y_test, y_test_pred_knn))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_knn))
plot_conf_matrix(y_test, y_test_pred_knn, 'K-Nearest Neighbors (k=1)')

# Trying with a higher value of k to reduce overfitting
knn = KNeighborsClassifier(n_neighbors=5)  # You can experiment with different values of k
knn.fit(X_train, y_train)

# Predictions on training data
y_train_pred_knn = knn.predict(X_train)
# Predictions on test data
y_test_pred_knn = knn.predict(X_test)

# Evaluating on training data
print("KNN Classifier (k=5) Report on Training Data:\n", classification_report(y_train, y_train_pred_knn))
print("Training Accuracy:", accuracy_score(y_train, y_train_pred_knn))

# Evaluating on test data
print("KNN Classifier (k=5) Report on Test Data:\n", classification_report(y_test, y_test_pred_knn))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred_knn))
plot_conf_matrix(y_test, y_test_pred_knn, 'K-Nearest Neighbors (k=5)')
````
